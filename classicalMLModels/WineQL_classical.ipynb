Tests whether a data sample has a Normal distribution.

Assumptions

Observations in each sample are independent and identically distributed (iid). Interpretation

H0: the sample has a Gaussian distribution.
H1: the sample does not have a Gaussian distribution.

```python
from scipy.stats import shapiro
features = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
for feature in features:
    stat, p = shapiro(df[feature])
    print('Feature {}'.format(feature))
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably feature "{}" is Normally Distributed'.format(feature))
        print('\n')
    else:
        print('Probably feature "{}" is not Normally Distributed'.format(feature))
        print('\n')
```

* Based on Shapiro-Wilk test: none of the features follow the normal distribution
* Normalization using min max scaler or power transformation would be useful prior to feeding the model.

# Summary


* The data is not balanced

# DATA PREP


```python
def new_quality_rating(rating):
    '''This function will change the target quality rating to 0, 1 or 2.
       The number of target labels will be reduced to 3 instead of 6'''
    if (rating == 3 or rating == 4):
        return 0
    elif (rating == 5 or rating == 6):
        return 1
    else:
        return 2
```

```python
df['new_quality'] = df['quality'].apply(new_quality_rating)
```

```python
df.head()
```

```python
new_df = df.copy()
```

```python
new_df.drop('quality',axis=1,inplace =True)
```

```python
new_df.head()
```

# Split data: 1 for training 1 for testing


```python
train,test = train_test_split(new_df,test_size=0.2,random_state=42,stratify=new_df['new_quality'])
```

```python
train.shape
# 3918 rows of observation
```

```python
test.shape
```

# Reason for stratify


```python
train['new_quality'].value_counts()/len(train)
```

```python
test['new_quality'].value_counts()/len(test)
```

```python
#separate for the train set
X_train = train.drop('new_quality',axis=1)
y_train = train['new_quality']

#separate for the test set
X_test = test.drop('new_quality',axis=1)
y_test = test['new_quality']
```

```python
#instantiate a standard scaler object
scaler = StandardScaler()

#fit and transform the train set
X_train = scaler.fit_transform(X_train)
#transform the test set
X_test = scaler.transform(X_test)
```

```python
# Apply power transformation to make the features more gaussian type
pt = PowerTransformer(method='yeo-johnson')

#power transform the train data
X_train = pt.fit_transform(X_train)
#power transform the test data
X_test = pt.transform(X_test)
```

# 1 use all features to train

```python
#instantiate the model object
clf = LogisticRegression(multi_class='multinomial',solver='lbfgs')
clf.fit(X_train,y_train)
```

```python
#predict on the X_train
train_predictions = clf.predict(X_train)
#predict on the X_test
test_predictions = clf.predict(X_test)
```

```python
#plot of the confusion matrix to show the predicted label vs the true label
cm = confusion_matrix(y_test,test_predictions)
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test,test_predictions,zero_division=0))
```

* overall accuracy 77%
* the accuracy for class 0: 0, class1: 94 class 2: 33
=> the model is not performing well with the less frequent class: 0 and 2.
maybe bcs the data is imbalanced


# 2 Apply Dimension reductionality

```python
#print the classification report which is a summary of all key classification metrics
print(classification_report(y_test,test_predictions,zero_division=0))
```

# ensemble-based hybrid model

```python
clf1 = LogisticRegression(multi_class='multinomial', solver='lbfgs')
clf2 = DecisionTreeClassifier(random_state=42)
clf3 = RandomForestClassifier(random_state=42)
```

```python
voting_clf = VotingClassifier(estimators=[('lr', clf1), ('dt', clf2), ('rf', clf3)], voting='soft')
```

```python
voting_clf.fit(X_train, y_train)
y_pred = voting_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
